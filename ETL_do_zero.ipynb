```python
import pandas as pd
import numpy as np
```

```python
#criar uma base de dados com 50 linhas e 5 colunas
np.random.seed(42)
data = {
    'ID': range(1, 51),
    'Nome': ['Pessoa' + str(i) for i in range(1, 51)],
    'Idade': np.random.randint(20, 60, size = 50),
    'Salario': np.random.randint(30000,100000, size = 50),
    'Departamento': np.random.choice(['Vendas', 'Marketing', 'TI', 'RH'], size = 50)
}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
#Renomear as colunas
df.columns = ['ID', 'Nome', 'Idade', 'Salario', 'Departamento']
#normatizar a coluna salario
df['Salario Normatizado'] = df['Salario']
df['Salario Normatizado'] = (df['Salario Normatizado'] - df['Salario Normatizado'].min()) / (df['Salario Normatizado'].max() - df['Salario Normatizado'].min())
```

```python
#Filtrar apenas pessoas com idade maior que 30
df_filtrado = df[df['Idade'] > 30]
```

```python
df_filtrado.head()
```

```python
# Salvar os dados transformados em um arquivo csv
df.to_csv('dados_transformados.csv', index = False)
```

```python
#Transformar a coluna salario em reais, no padrão brasileiro
df['Salario Normatizado'] = 'RS' + df['Salario Normatizado'].astype(float).map('{:,.2f}'.format).str.replace(',', ' ')
```
